```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/nishant-harinkhede12/EDA-Projects/main/country_vaccinations.csv'
data = pd.read_csv(url,parse_dates=True)
data.head()
```

```python
data.country.value_counts()
```

```python
data = data[data.country.apply(lambda x: x not in ["England", "Scotland", "Wales", "Northern Ireland"])]
data.country.value_counts()
```

```python
df = data[['country','vaccines']]
```

Now letâ€™s see how many countries are taking each of the vaccines mentioned in this data:

```python
dict_ = {}
for i in df.vaccines.unique():
  dict_[i] = [df["country"][j] for j in df[df["vaccines"]==i].index]

vaccines = {}
for key, value in dict_.items():
  vaccines[key] = set(value)
for i, j in vaccines.items():
  print(f"{i}:>>{j}")
```

```python
import plotly.express as px
import plotly.offline as py

vaccine_map = px.choropleth(data, locations = 'iso_code', color = 'daily_vaccinations_per_million')
vaccine_map.update_layout(height=300, margin={"r":0,"t":0,"l":0,"b":0})
vaccine_map.show()
```

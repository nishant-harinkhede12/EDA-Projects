```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

url = 'https://raw.githubusercontent.com/nishant-harinkhede12/EDA-Projects/main/loan_data.csv'

data = pd.read_csv(url)
data.head()
```

```python
data['credit.policy'].value_counts()
```

```python
plt.hist(data[data['credit.policy']==1]['fico'],bins=30,alpha=0.5,color='blue', label='Credit.Policy=1')
plt.hist(data[data['credit.policy']==0]['fico'],bins=30,alpha=0.5,color='red', label='Credit.Policy=0')
plt.legend()
plt.xlabel('FICO Score')
plt.title('....Jai Mata Di....')
plt.show()
# data[data['credit.policy']==1]['fico'].plot.hist(bins=30,alpha=0.5,color='blue', label='Credit.Policy=1')
```

```python
sns.boxplot(x=data['credit.policy'],y=data['int.rate'])
plt.title('Interest rate varies between risky and non-risky borrowers')
plt.show()
```

```python
sns.boxplot(x=data['credit.policy'],y=data['log.annual.inc'])
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(x=data['purpose'],hue=data['not.fully.paid'], data = data,palette='Set1')

plt.title("Bar chart of loan purpose colored by not fully paid status", fontsize=17)
plt.xlabel("Purpose", fontsize=15)
```

```python
sns.jointplot(x=data['fico'],y=data['int.rate'],data=data,color='purple', size = 18)
```

```python

plt.figure(figsize=(10,5))
sns.lmplot(x='fico',y='int.rate',data=data, hue='credit.policy',col='not.fully.paid', palette='Set1',size=5)
```
